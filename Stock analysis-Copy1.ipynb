```python
start= datetime.datetime(2012,1,1)
end = datetime.datetime(2017,1,1)
```

```python
tesla= web.DataReader("TSLA",'yahoo',start,end)
ford= web.DataReader("F",'yahoo',start,end)
gm= web.DataReader("GM",'yahoo',start,end)
```

```python
tesla.to_csv('Tesla_Stock.csv')
ford.to_csv('Ford_Stock.csv')
gm.to_csv('GM_Stock.csv')
```

```python
tesla.head()
```

```python
ford.tail()
```

```python
gm.head()
```

```python
tesla['Open'].plot(label="Tesla",figsize=(15,8))
gm['Open'].plot(label="GM")
ford['Open'].plot(label="Ford")
plt.legend()
plt.show()
```

```python
tesla['Volume'].plot(label="Tesla",figsize=(15,8))
gm['Volume'].plot(label="GM")
ford['Volume'].plot(label="Ford")
plt.ylabel('Volume Traded')
plt.legend()
plt.show()
```

```python
ford.iloc[470:550]['Open'].plot()
```

```python
tesla['Total Traded']=tesla['Open']*tesla['Volume']
gm['Total Traded']=gm['Open']*gm['Volume']
ford['Total Traded']=ford['Open']*ford['Volume']
```

```python
tesla.head()
```

```python
ford.head()
```

```python
tesla['Total Traded'].plot(label='Tesla',figsize=(15,7))
gm['Total Traded'].plot(label='GM')
ford['Total Traded'].plot(label='Ford')
plt.legend()
```

```python
gm['Open'].plot(label='No Moving Average',figsize=(15,7))
gm['MA20']=gm['Open'].rolling(20).mean()
gm['MA20'].plot(label='MA20')
gm['MA200']=gm['Open'].rolling(200).mean()
gm['MA200'].plot(label='MA200')
plt.legend()
```

```python
short_rolling = gm.rolling(window=20).mean()
short_rolling.head(20)

ema_short = gm.ewm(span=20, adjust=False).mean()

fig, ax = plt.subplots(figsize=(15,9))

ax.plot(gm.loc[start:end, :].index, gm.loc[start:end, 'Open'], label='Price')
ax.plot(ema_short.loc[start:end, :].index, ema_short.loc[start:end, 'Open'], label = 'Span 20-days EMA')
ax.plot(short_rolling.loc[start:end, :].index, short_rolling.loc[start:end, 'Open'], label = '20-days SMA')

ax.legend(loc='best')
ax.set_ylabel('Price in $')
```
